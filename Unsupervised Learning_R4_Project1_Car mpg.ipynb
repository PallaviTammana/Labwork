# <span style="font-family:  Lucida Calligraphy; font-size:1.2em;color:navy;"> Outliers Treatment: 
    1. We know that we have outliers in data, correct them (replace with median values)
    2. rebuild the clusters and check if we get any better results

```python
def fixoutliers(df):
    col_names=df.columns
    for col in col_names:
        try:
            col_data=df[col]    
            mydata=pd.DataFrame()
            updated=[]
            Q1,Q2,Q3=np.percentile(col_data,[25,50,75])
            IQR=Q3-Q1
            minimum=Q1-1.5*IQR
            maximum=Q3+1.5*IQR
            for row_val in col_data:
                if(row_val>maximum):
                    row_val=Q2
                    updated.append(row_val)
                elif(row_val<minimum):
                    row_val=Q2
                    updated.append(row_val)
                else:
                    updated.append(row_val)
            df[col]=updated
        except:
            continue
    return df
# KNN model with outliers corrected
KM3_no_outliers=cars_z.copy()
KM3_no_outliers=fixoutliers(KM3_no_outliers)
cluster = KMeans( n_clusters = 3, random_state = 1 )
cluster.fit(KM3_no_outliers)
print((metrics.silhouette_score(KM3_no_outliers, labels, metric='euclidean')))
centroids = cluster.cluster_centers_
centroid_df = pd.DataFrame(centroids, columns = list(KM3_no_outliers) )
prediction=cluster.predict(KM3_no_outliers)
KM3_no_outliers["Label"] = prediction
KM3_no_outliers.groupby("Label").count()
```

# <span style="font-family:  Lucida Calligraphy; font-size:1.2em;color:navy;"> Linear Model Building: 
    1. Build all linear models on full data with outliers
    2. Build all linear models on full data without outliers
    3. find the best model in each group

```python
def plot_perf(df):
    def output_tracer(metric,color) :
        tracer = go.Bar(y = df["Model_Used"] +' model on '+df["Data_Used"],
                        x = df[metric],
                        orientation = "h",
                        name = metric ,
                        textposition='auto',
                        marker = dict(line = dict(width =.7),color = color)
                       )
        return tracer
    layout = go.Layout(dict(title = " Model performances",
                            plot_bgcolor  = "rgb(243,243,243)",
                            paper_bgcolor = "rgb(243,243,243)",
                            xaxis = dict(gridcolor = 'rgb(255, 255, 255)',
                                         title = "R2 Score",
                                         zerolinewidth=1,
                                         ticklen=5,gridwidth=2),
                            yaxis = dict(gridcolor = 'rgb(255, 255, 255)',
                                         zerolinewidth=1,ticklen=5,gridwidth=2),
                            margin = dict(l = 250),
                            height = 780
                           )
                      )
    trace1  = output_tracer("Training_Score","#6699FF")
    trace2  = output_tracer('Testing_Score',"#33CC99")
    data = [trace1,trace2]
    fig = go.Figure(data=data,layout=layout)
    py.iplot(fig)

def get_best_model_per_group(df):
    groups=df.groupby(by='Group',axis=0)['Testing_Score'].max().keys()
    scores=df.groupby(by='Group',axis=0)['Testing_Score'].max().values
    return_df= pd.DataFrame(columns=df.columns)
    for cnt in range(len(groups)):
        dummy= df[(df['Testing_Score']==scores[cnt]) & (df['Group']==groups[cnt])]
        return_df = pd.concat([return_df,dummy],axis = 0).reset_index()  
        return_df = return_df.drop(columns = "index",axis =1)
    plot_perf(return_df)
    return return_df

def get_model_score(model,model_name,df_name,X,y,group_number):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=1)
    model.fit(X_train, y_train)
    s_df = pd.DataFrame({"Model_Used"           : [model_name],
                         "Group":[group_number],
                       "Data_Used"        : [df_name],
                       "Training_Score": [(model.score(X_train, y_train))*100],
                       "Testing_Score"    : [(model.score(X_test, y_test))*100],
                      })
    return s_df

    
def find_best_model(df,df_name,group_number):
    X=df.drop('mpg',axis=1)
    y=df['mpg']
    
    regression_model = LinearRegression()
    reg_stats=get_model_score(regression_model,'Linear',df_name,X,y,group_number)
    
    ridge = Ridge(alpha=.3,random_state=1)
    ridge_stats=get_model_score(ridge,'Ridge',df_name,X,y,group_number)
    
    lasso = Lasso(alpha=0.1,random_state=1)
    lasso_stats=get_model_score(lasso,'Lasso',df_name,X,y,group_number)
    
    performances = pd.concat([reg_stats,ridge_stats,lasso_stats,],axis = 0).reset_index()  
    performances = performances.drop(columns = "index",axis =1)
    
#     best_perf_model = get_best_model_per_group(performances)
    plot_perf(performances)
    return performances
```
